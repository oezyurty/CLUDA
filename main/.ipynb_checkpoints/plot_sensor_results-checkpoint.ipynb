```python
# Fixing random state for reproducibility
np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.barh(y_pos, performance, left=performance, align='center', color="red")
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

```python
def autolabel(rects):
    """
    Attach a text label in the middle of that bar displaying its value
    """
    for rect in rects:        
        if rect.get_x() == 0:
            val = rect.get_width()
            ax.text(rect.get_x() + rect.get_width()/2., rect.get_y() + rect.get_height()/2.,
                    '%.1f' % val,
                    ha='center', va='center')
        
        else:
            val = rect.get_width()
            ax.text(rect.get_x() + rect.get_width()/2., rect.get_y() + rect.get_height()/2.,
                    '+%.1f' % val,
                    ha='center', va='center')
```

```python
# Accuracy Results
plt.rcdefaults()
fig, ax = plt.subplots(figsize=(15,5))

TCN = np.array([63.9, 65.5, 49.6])
best_base = np.array([66.9, 79.4, 62.3])
cluda = np.array([75.4, 94.4, 75.9])

algos = ["w/o UDA", "Best Baseline", "CLUDA"]
y_pos = np.arange(len(algos))

rect_TCN = ax.barh(y_pos, TCN, align='center', color="tab:blue")
rect_best_base = ax.barh(y_pos, best_base-TCN, left=TCN, align='center', color="tab:orange")
rect_cluda = ax.barh(y_pos, cluda-best_base, left=best_base, align='center', color="tab:green")

ax.set_yticks(y_pos)
ax.set_yticklabels(algos)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Accuracy (%)')

autolabel(rect_TCN)
autolabel(rect_best_base)
autolabel(rect_cluda)

plt.show()
```
